```python
import pandas as pd

train = pd.read_csv("data/train.csv")
train.head()
```

```python
train_id = train['id']
train_lab = train['species']
```

```python
train.drop(['id','species'],inplace=True,axis=1)
train.head()
```

```python
train.describe()
```

```python
%matplotlib inline

import matplotlib
matplotlib.style.use('ggplot')
import matplotlib.pyplot as plt

train.boxplot(return_type='axes', figsize=(20, 4))
```

```python
import seaborn as sns
import re
sns.set_style('ticks')
fig, ax = plt.subplots()
# the size of A4 paper
fig.set_size_inches(12,8)
vnames = list(train)
vnames = [re.compile('[^a-zA-Z]').sub('', x) for x in vnames]
vnames2 = list(vnames)
for i, item in enumerate(vnames):
    if i ==0: next
    if item == vnames[i-1]:
        vnames2[i]=''
axobj = sns.violinplot(data=train, inner="points", ax=ax, )    
sns.despine()
axobj.set_xticklabels(vnames2)
```

There is something about the shape features. Let's take a closer look.

```python
train_shape = train.filter(regex='shape')
train_shape.describe()
```

```python
%matplotlib inline
sns.set_style('ticks')
fig, ax = plt.subplots()
# the size of A4 paper
fig.set_size_inches(12,8)
vnames = list(train_shape)
vnames = [re.compile('[^a-zA-Z]').sub('', x) for x in vnames]
vnames2 = list(vnames)
for i, item in enumerate(vnames):
    if i ==0: next
    if item == vnames[i-1]:
        vnames2[i]=''
axobj = sns.violinplot(data=train_shape, inner="points", ax=ax, )    
sns.despine()
axobj.set_xticklabels(vnames2)
```

Well this is a lovely sinusoidal pattern. What does it mean?

```python
from sklearn.decomposition import PCA
from sklearn import preprocessing
```

```python
scaler = preprocessing.StandardScaler().fit(train)
pca = PCA()
print(scaler.transform(train).std(axis=0))
pca.fit(scaler.transform(train)  )
```

```python
print(pca.explained_variance_ratio_.shape) 
print(train.shape)
```

```python
expvar = pd.DataFrame(dict(y=pca.explained_variance_ratio_,factor=list(range(192))))

ax = sns.barplot(y="y", x="factor", data=expvar[0:50])
plt.setp(ax.get_xticklabels(), rotation=45)
```
